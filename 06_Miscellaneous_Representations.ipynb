```python
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
```

```python
file = 'data/WatkinsPMacro.wav'
wf, sr_sw = sf.read(file)
signal_sw = wf[31000:37500]
time_sw = np.linspace(0,len(signal_sw)/sr_sw,len(signal_sw))
signal_sw[0], signal_sw[-1] = 0, 0
N = len(signal_sw)
```

```python
def find_nearest(arr, val):
    arr = np.asarray(arr)
    idx = (np.abs(arr - val)).argmin()
    return idx, arr[idx]
```

```python
func = lambda tau, t: signal_sw[find_nearest(time_sw, t+tau/2)[0]] * np.conj(signal_sw[find_nearest(time_sw, t-tau/2)[0]])
```

```python
t_res = 256
f_res = 256
ts = np.linspace(0, time_sw[-1], t_res)
fs = np.linspace(0, 1, f_res)
mat = np.zeros((t_res, f_res), dtype=complex)
for i, t in enumerate(ts):
    taus = np.linspace(max(-2*t, -2*(1-t)), min(2*t, 2*(1-t)), f_res)

    funcs = [func(tau, t) for tau in taus]
    x = np.fft.fft(funcs)
    mat[i, :] = x

ts, fs = np.meshgrid(ts, fs)
plt.pcolormesh(ts, fs, np.log10(np.abs(np.real(mat.T))+1e-2))
plt.colorbar()
plt.show()
```
